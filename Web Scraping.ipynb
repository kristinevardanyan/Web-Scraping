### Step 5: Scraping Data from Product Page

```python
watches_db = list()
count = 0

for watch in urls_to_csv['URLs'].tolist()[:200]:
    webpage = requests.get(watch)
    soup = BeautifulSoup(webpage.content, "html.parser")
    
    time.sleep(1)
    watches_db.append({
        'Product ID': soup.find('div', class_='product__features').find("h1").text,
        'Price (AMD)': soup.find('div', class_='product__price').text[:-1]
    })
    
    count += 1
    print('Observation:', count)

watches_df = pd.DataFrame(watches_db)
watches_df.to_csv('Casio_watches.csv', index=False)
```
